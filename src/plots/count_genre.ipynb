```python
import seaborn as sns
from matplotlib.font_manager import fontManager, FontProperties
from paths import TEST_OUTPUT_DIR
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path = "NotoSansTC-Medium.ttf"
fontManager.addfont(path)

prop = FontProperties(fname=path)
sns.set(font=prop.get_name())
```

```python
data = pd.read_csv(TEST_OUTPUT_DIR / "count_genre.csv")

data["1"] = data["1"] / sum(data["1"]) * 100
plt.figure(figsize=(10, 20), dpi=500)
fig = sns.barplot(data, x="1", y="0", legend="")
fig.axes.set_title("spotify-tracks-dataset中各音樂類型相對次數分配長條圖", fontsize=20)
fig.set_xlabel("相對次數 (%)", fontsize=16)
fig.set_ylabel("音樂類型", fontsize=16)
```
