```python
import seaborn as sns
from matplotlib.font_manager import fontManager, FontProperties
from paths import TEST_OUTPUT_DIR
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

```python
path = "NotoSansTC-Medium.ttf"
fontManager.addfont(path)

prop = FontProperties(fname=path)
sns.set(font=prop.get_name())
```

```python
data = pd.read_csv(TEST_OUTPUT_DIR / "blank_threshold.csv")

data["音檔數值低於臨界值的相對次數 (%)"] = data["音檔數值低於臨界值的相對次數 (%)"] * 100

plt.figure(figsize=(8, 5))
fig = sns.lineplot(data, x="臨界值 (librosa.load)", y="音檔數值低於臨界值的相對次數 (%)", hue="段落類別")
fig.axes.set_title("音檔數值累積相對次數分配折線圖")
plt.xticks(data["臨界值 (librosa.load)"][::4], rotation=45)
```
